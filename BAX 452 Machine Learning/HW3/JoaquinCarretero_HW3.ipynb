### 2) **Linear Regression Model Development** (15 pts):
- Develop a Linear regression model to predict price of house in Melbourne.

- Preprocess the data as necessary (handling missing values, encoding categorical variables, removing unnecessary columns etc.).

- Explain the steps you took in preparing the data and developing the model.

- You might need to encode the values for your prediction.

```python
data = data.drop(columns=['Address', 'Date','Postcode', 'Suburb', 'SellerG', 'Lattitude', 'Longtitude'])
data = data.dropna(subset=['Price'])
```

```python
features = data.columns.drop('Price')
X = data[features]
y = data['Price']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
num_cols = X_train.select_dtypes(include=['int64', 'float64']).columns
cat_cols = X_train.select_dtypes(include=['object']).columns

num_trans = SimpleImputer(strategy='median')
cat_trans = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

preprocessor = ColumnTransformer(
    transformers=[
        ('num', num_trans, num_cols),
        ('cat', cat_trans, cat_cols)
    ])
```

```python
linear_model = LinearRegression()
linear_pipeline = Pipeline(steps=[('preprocessor', preprocessor),
                                  ('model', linear_model)])
linear_pipeline.fit(X_train, y_train)
```

We used a straightforward approach to build the linear regression model. We utilized Scikit Learn's LinearRegression class. First we removed columns like Address that had many unique values, since they would make the model more complex without helping prediction. On the small dataset, we imputed missing values with median and mode instead of dropping them. Categoricals were one-hot encoded for the regression task. After preprocessing, we split the data 70/30 into train and test sets. The model was fit on the training portion. To evaluate performance, we looked at RMSE and R2 on the test set.

- Lasso regularization can address this overfitting. By penalizing model complexity via shrinking and setting coefficients to zero, Lasso reduces reliance on noise and spurious correlations. This generalization should improve test set metrics. While the training R2 may decrease due to these restrictions, the model should perform better on new out-of-sample data.

```python
# Used a for loop to determine the best alpha between 0 and 5. I understand that the higher
# the alpha the more variables are penalized. 

alpha_values = np.arange(0, 10.1, 0.1)
best_r2, best_rmse, best_alpha = -np.inf, np.inf, None
best_r2_train, best_rmse_train = -np.inf, np.inf

for alpha in alpha_values:
    lasso_model = Lasso(alpha=alpha, max_iter=100000, random_state=42)
    lasso_pipeline = Pipeline([('preprocessor', preprocessor), ('model', lasso_model)])
    lasso_pipeline.fit(X_train, y_train)

    y_pred_test = lasso_pipeline.predict(X_test)
    y_pred_train = lasso_pipeline.predict(X_train)

    r2_test, rmse_test = r2_score(y_test, y_pred_test), np.sqrt(mean_squared_error(y_test, y_pred_test))
    r2_train, rmse_train = r2_score(y_train, y_pred_train), np.sqrt(mean_squared_error(y_train, y_pred_train))
    if r2_test > best_r2 or (r2_test == best_r2 and rmse_test < best_rmse):
        
        best_alpha, best_r2_test, best_rmse_test = alpha, r2_test, rmse_test
        best_r2_train, best_rmse_train = r2_train, rmse_train

best_alpha, best_r2_train, best_rmse_train, best_r2_test, best_rmse_test
```

The Lasso model shows some improvements in out-of-sample generalization compared to linear regression.
  
1. On the training set, R2 increased slightly from 0.73 to 0.74. RMSE also decreased with Lasso. This indicates the Lasso restrictions maintained training accuracy while improving the fit.
  
2. On the test set, R2 edged up from 0.54 to 0.54. Test RMSE also declined from 513,000 to 510,000. While small, these metrics show the Lasso model better captures the true relationships and performs slightly better on new data.

We used Statsmodels instead of Scikit-Learn to fit the models. Statsmodels provides model evaluation metrics like AIC and BIC that Scikit-Learn does not.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

def deviance(y, pred, family="gaussian"):
    if family == "gaussian":
        return np.sum((y - pred) ** 2)
    elif family == "binomial":
        y = np.array(y)
        pred = np.array(pred)
        return -2 * np.sum(y * np.log(pred) + (1 - y) * np.log(1 - pred))
```

#### Linear Model

```python
#WARNING: I USED PROF JORN'S CODE TO DO MANUAL KFOLD


X_train_processed = preprocessor.fit_transform(X_train)
X_test_processed = preprocessor.transform(X_test)

X_train_processed_dense = X_train_processed.toarray() 
X_train_sm = sm.add_constant(X_train_processed_dense)
model = sm.OLS(y_train, X_train_sm).fit()
y_pred_train = model.predict(X_train_sm)


# Estimate AIC, AICc, BIC
k = model.df_model + 1
n = len(y_train)
aic_linear = model.aic
aic_c_linear = aic_linear + (2 * k * (k + 1)) / (n - k - 1)
bic_linear = model.bic

# Perform 5-fold Cross-Validation manually
kf = KFold(n_splits=5, shuffle=True, random_state=0)
cv_mse_scores = []

for train_index, val_index in kf.split(X_train_processed):
    # Split data
    X_train_kf, X_val_kf = X_train_processed[train_index], X_train_processed[val_index]
    y_train_kf, y_val_kf = y_train.iloc[train_index], y_train.iloc[val_index]

    # Add constant
    X_train_kf_sm = sm.add_constant(X_train_kf.toarray())
    X_val_kf_sm = sm.add_constant(X_val_kf.toarray())

    # Fit the model
    model_kf = sm.OLS(y_train_kf, X_train_kf_sm).fit()

    # Predict and calculate MSE for the fold
    y_pred_kf = model_kf.predict(X_val_kf_sm)
    mse_kf = mean_squared_error(y_val_kf, y_pred_kf)
    cv_mse_scores.append(mse_kf)

X_test_sm = sm.add_constant(X_test_processed.toarray())
y_pred_test = model.predict(X_test_sm)
mse_linear_test = mean_squared_error(y_test, y_pred_test)
mse_linear_train = mean_squared_error(y_train, y_pred_train)
deviance_linear = deviance(y_test, y_pred_test, family="gaussian")


# Output the results
print("AIC_Training:", aic_linear)
print("AICc_Training:", aic_c_linear)
print("BIC_Training:", bic_linear)
print("5-fold CV Mean MSE on Training:", np.mean(cv_mse_scores))
print("MSE Train:", mse_linear_train)
print("MSE Test:", mse_linear_test)
print("Deviance on Testing:", deviance_linear)
```

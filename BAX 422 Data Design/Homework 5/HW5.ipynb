```python
count = 1
contributors_details = []
print(contributors)

for contributor in contributors:
    username = contributor['login']
    url = f'https://api.github.com/users/{username}'
    user_details = requests.get(url, headers=headers).json()
    if user_details:
        # Extract and print the required information
        print(f"User {count} Details")
        print("Login: ", user_details.get('login'))
        print("ID: ", user_details.get('id'))
        print("Location: ", user_details.get('location'))
        print("Email: ", user_details.get('email'))
        print("Hireable: ", user_details.get('hireable'))
        print("Bio: ", user_details.get('bio'))
        print("Twitter Username: ", user_details.get('twitter_username'))
        print("Public Repos: ", user_details.get('public_repos'))
        print("Public Gists: ", user_details.get('public_gists'))
        print("Followers: ", user_details.get('followers'))
        print("Following: ", user_details.get('following'))
        print("Created At: ", user_details.get('created_at'))
        print("\n")
        
        # Append the user details to the list
        contributors_details.append(user_details)
    count += 1
```

```python
import mysql.connector
from mysql.connector import Error

connection = mysql.connector.connect(
    host='localhost',       
    user='root',  
    password=''
)

cursor = connection.cursor()

# Drop the database if it exists and then create it
cursor.execute("DROP DATABASE IF EXISTS contributors_db")
cursor.execute("CREATE DATABASE IF NOT EXISTS contributors_db")
cursor.execute("USE contributors_db")

# Create the contributors table
cursor.execute('''
CREATE TABLE IF NOT EXISTS contributors (
    id INT PRIMARY KEY,
    login VARCHAR(255) NOT NULL,
    location VARCHAR(255),
    email VARCHAR(255),
    hireable BOOLEAN,
    bio TEXT,
    twitter_username VARCHAR(255),
    public_repos INT,
    public_gists INT,
    followers INT,
    following INT,
    created_at VARCHAR(255)
)
''')

# Creating indexes for optimized lookups
cursor.execute("CREATE INDEX idx_login ON contributors(login)")
cursor.execute("CREATE INDEX idx_location ON contributors(location)")
cursor.execute("CREATE INDEX idx_hireable ON contributors(hireable)")

# Insert contributors' details into the database
insert_query = '''
INSERT INTO contributors (id, login, location, email, hireable, bio, twitter_username, public_repos, public_gists, followers, following, created_at)
VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
'''
for details in contributors_details:
    cursor.execute(insert_query, (
        details['id'], details['login'], details['location'], details['email'],
        details['hireable'], details['bio'], details['twitter_username'],
        details['public_repos'], details['public_gists'], details['followers'],
        details['following'], details['created_at']
    ))

# Test and print the inserted records
cursor.execute("SELECT * FROM contributors")
print("All contributors:", cursor.fetchall())

# Test and print the records with location as Tokyo (returns an empty list)
cursor.execute("SELECT * FROM contributors WHERE location = 'Tokyo'")
print("Tokyo location:", cursor.fetchall())

# Test and print the records with location as Bangalore (returns a lengthy list)
cursor.execute("SELECT * FROM contributors WHERE location = 'Bangalore'")
print("Bangalore location:", cursor.fetchall())

# Commit changes
connection.commit()
print("Records inserted successfully")

cursor.close()
connection.close()

print("MySQL connection is closed")
```

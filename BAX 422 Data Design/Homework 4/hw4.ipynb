## Questuin 1

1. Navigate to google.com and search "askew"
   
2. Do the same with "google in 1998"
   
   

```python
from webdriver_manager.chrome import ChromeDriverManager
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait
import time
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup

# Web scraping with Selenium
options = webdriver.ChromeOptions()
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

# Navigate to google.com and search "askew"
driver.get("https://www.google.com")
search_box = driver.find_element(By.NAME, "q")
search_box.clear()
search_box.send_keys("askew")
search_box.send_keys(Keys.RETURN)  # Press Enter to search
time.sleep(3)

# Navigate to google.com and search "google in 1998"
driver.get("https://www.google.com")
search_box = driver.find_element(By.NAME, "q")
search_box.clear()
search_box.send_keys("google in 1998")
search_box.send_keys(Keys.RETURN)  # Press Enter to search
time.sleep(3)

# Close the browser
driver.quit()
```

## Question 2

1. Go to bestbuy.com
2. Click on Deals of the Day
3. Read how much time is left for the Deal of the Day and print the remaining time
4. Click on the Deal of the Day (actual product)
5. Click on Reviews
6. Save html as bestbuy_deal_of_the_day.html

```python
# 1. Go to bestbuy.com
options = webdriver.ChromeOptions()
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)
driver.get("https://www.bestbuy.com")

# 2. Click on Deal of the Day
deals_of_the_day = driver.find_element(By.LINK_TEXT, "Deal of the Day")
deals_of_the_day.click()

# 3. Read how much time is left for the Deal of the Day and print the remaining time
print(driver.find_element(By.ID, "countdown").text.replace('\n', ' ').strip())

# 4. Click on the Deal of the Day (actual product)
deal_of_the_day = driver.find_element(By.CLASS_NAME, "wf-offer-link")
deal_of_the_day.click()

# 5. Click on link containing "User rating" as part of the text
user_rating = driver.find_element(By.PARTIAL_LINK_TEXT, "User rating")
user_rating.click()

time.sleep(5)
# 6. Save html as bestbuy_deal_of_the_day.html
with open('bestbuy_deal_of_the_day.html', 'w') as f:
    f.write(driver.page_source)

# 7. Close the browser
driver.quit()
```

## Question 3

1. Go to lavazzausa.com rejecting all cookies
2. Click on Product
3. Click on Capsule coffee machines
4. Save page to lavazza.html
5. Find the original price of the "Classy Plus Bundle" and print it using BeautifulSoup

```python
# 1. Go to lavazzausa.com 
options = webdriver.ChromeOptions()
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)
driver.get("https://www.lavazzausa.com")

# 2. Reject all cookies id="consent_prompt_reject"
time.sleep(3)
reject_all_cookies = driver.find_element(By.ID, "consent_prompt_reject")
reject_all_cookies.click()

# 3. Click on Product
products = driver.find_element(By.ID, "PRODUCTS")
products.click()

# 4. Click on "Capsule coffee machines"
capsule_coffee_machines = driver.find_element(By.LINK_TEXT, "Capsule coffee machines")
capsule_coffee_machines.click()

# 5. Save page to lavazza.html
with open('lavazza.html', 'w') as f:
    f.write(driver.page_source)

# 6. Using BeautifulSoup, read lavvazza.html and print the price of Classy Plus Bundle
with open('lavazza.html', 'r') as f:
    soup = BeautifulSoup(f, 'html.parser')

# Find the article by data-product-name and extract the 'data-price' attribute
article = soup.find('article', {'data-product-name': "Classy Plus Bundle"})
price = article['data-price']
print(f'The price is: {price}')

# 7. Close the browser
driver.quit()
```
